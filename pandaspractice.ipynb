```python
# prompt: The following table gives the count of matched strings and count of
# mismatch on each coordinate. With the help of python first create a
# dictionary with the data provided. Next create a csv file with the help of a
# dictionary (create it according to the given table). Next calculate the standard
# score for each Count value using the expression (x- mean)/stddev.                                                                          Coordinate Count Mismatch
# C1 300 5
# C2 210 13
# C3 230 14
# C4 260 3
# C5 270 2

import pandas as pd
import numpy as np

# Create the dictionary
data = {
    'Coordinate': ['C1', 'C2', 'C3', 'C4', 'C5'],
    'Count': [300, 210, 230, 260, 270],
    'Mismatch': [5, 13, 14, 3, 2]
}

# Create a pandas DataFrame from the dictionary
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv('coordinate_data.csv', index=False)

# Calculate the mean and standard deviation of the 'Count' column
count_mean = df['Count'].mean()
count_std = df['Count'].std()

# Calculate the standard score for each 'Count' value
df['Standard_Score'] = (df['Count'] - count_mean) / count_std

# Print the DataFrame with the standard scores
df
```

```python
# prompt: Take a paragraph of text input. Write the contents in a file r1.txt. Open
# the file again and read and print its content. Copy the contents of the file in
# another file r2.txt, and print its content.

paragraph = input("Enter a paragraph of text: ")

# Write the paragraph to r1.txt
with open("r1.txt", "w") as file:
  file.write(paragraph)

# Read and print the content of r1.txt
with open("r1.txt", "r") as file:
  content = file.read()
  print("Content of r1.txt:\n", content)

# Copy the content of r1.txt to r2.txt
with open("r1.txt", "r") as file1, open("r2.txt", "w") as file2:
  file2.write(file1.read())

# Read and print the content of r2.txt
with open("r2.txt", "r") as file:
  content = file.read()
  print("\nContent of r2.txt:\n", content)
```

```python
# prompt: Consider n random numbers in the range 60 to 90. Write the numbers in
# a file. Next open the file in read mode and display its content, also display
# the maximum and minimum value.

import random
import pandas as pd

# Generate n random numbers between 60 and 90
n = 10  # You can change the number of random numbers
random_numbers = [random.randint(60, 90) for _ in range(n)]

# Write the numbers to a file
with open("random_numbers.txt", "w") as f:
    for number in random_numbers:
        f.write(str(number) + "\n")

# Open the file in read mode and display its content
with open("random_numbers.txt", "r") as f:
    print("Content of the file:")
    file_contents = f.readlines()
    for line in file_contents:
      print(line.strip()) #removes extra space

# Find the maximum and minimum values
numbers = [int(line.strip()) for line in file_contents] #convert strings to numbers
maximum = max(numbers)
minimum = min(numbers)

print("\nMaximum value:", maximum)
print("Minimum value:", minimum)
```

```python
# prompt: With respect to the world, assuming the world has n people in all, in
# the year 2019 the number of COVID effected patients were 20% of the
# total population, the number of effected people were 60% in the year 2020
# and 80% were affected in the year 2021. Prepare a list reflecting the actual
# values. Display the list. Prepare a pie chart with respect to the following
# data. Hint: Use the pie method as Plt.pie(data, labels=labels,
# autopct=’%.2f’) Here labels is a list of years for which you are
# determining the percentage of people effected. every percentage values are
# converted to 2 digit precisions.

import matplotlib.pyplot as plt

n = 1000  # Total population
covid_2019 = int(0.20 * n)
covid_2020 = int(0.60 * n)
covid_2021 = int(0.80 * n)

covid_data = [covid_2019, covid_2020, covid_2021]
years = ['2019', '2020', '2021']

print(covid_data)

plt.pie(covid_data, labels=years, autopct='%.2f%%')
plt.title('COVID Affected Population Percentage')
plt.show()
```

```python
# prompt: Using Pandas read a csv file from the following
# url https://media.geeksforgeeks.org/wp-content/uploads/nba.csv
# a) Print the size, shape, and dimension of the data frame created.
# b) Load the IRIS dataset. Preview the data present. Display the
# features available. Display the class level.
# c) Create a Data frame with the above data and replace
# the class level 0 , 1 and 2 with ‘Setosa’,‘Versicolor’, and
# ‘Virginica’
# d) Next, group the class level column based on the Species.
# e) Print the mean of sepal length and sepal width of each group
# f) Print the max, standard deviation value of each feature

import pandas as pd
import numpy as np

# a)
url = "https://media.geeksforgeeks.org/wp-content/uploads/nba.csv"
nba_df = pd.read_csv(url)

print("Size:", nba_df.size)
print("Shape:", nba_df.shape)
print("Dimension:", nba_df.ndim)

# b)
iris_df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None)
iris_df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']

print("\nIris Data Preview:")
print(iris_df.head())
print("\nIris Features:", iris_df.columns[:-1].tolist())  # Exclude the class column
print("\nIris Class Levels:", iris_df['class'].unique())

# c)
iris_df['class'] = iris_df['class'].replace({'Iris-setosa': 'Setosa', 'Iris-versicolor': 'Versicolor', 'Iris-virginica': 'Virginica'})

# d)
grouped_data = iris_df.groupby('class')


# e)
print("\nMean of sepal length and sepal width for each group:")
print(grouped_data[['sepal_length', 'sepal_width']].mean())

# f)
print("\nMax, Standard Deviation of each feature:")
print(grouped_data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].agg(['max', 'std']))
```

```python
# prompt: Using sklearn library cluster the following dataset into 3 clusters
# by kmeans algorithm. First create a csv file with the dataset.
# Next read the file and plot the data points considering the latitude and
# longitude features. Thereafter use kmeans and plot the clusters.
# Country Latitude Longitude Language
# USA 44.97 -103.77 English
# Canada 62.40 -96.80 English
# France 46.75 2.40 French
# UK 54.01 -2.53 English
# Germany 51.15 10.40 German
# Australia -25.45 133.11 English

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Create a DataFrame with the provided data
data = {
    'Country': ['USA', 'Canada', 'France', 'UK', 'Germany', 'Australia'],
    'Latitude': [44.97, 62.40, 46.75, 54.01, 51.15, -25.45],
    'Longitude': [-103.77, -96.80, 2.40, -2.53, 10.40, 133.11],
    'Language': ['English', 'English', 'French', 'English', 'German', 'English']
}
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv('country_data.csv', index=False)

# Read the data from the CSV file
df = pd.read_csv('country_data.csv')

# Plot the data points
plt.figure(figsize=(8, 6))
plt.scatter(df['Longitude'], df['Latitude'])
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Country Locations')
plt.show()

# Select features for clustering
X = df[['Latitude', 'Longitude']]

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=0)
df['Cluster'] = kmeans.fit_predict(X)

# Plot the clusters
plt.figure(figsize=(8, 6))
for cluster in range(3):
    cluster_data = df[df['Cluster'] == cluster]
    plt.scatter(cluster_data['Longitude'], cluster_data['Latitude'], label=f'Cluster {cluster}')
plt.scatter(kmeans.cluster_centers_[:, 1], kmeans.cluster_centers_[:, 0], s=200, c='red', marker='*', label='Centroids')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('KMeans Clustering of Countries')
plt.legend()
plt.show()
```
