Write a program to plot the following functions in different graph but in the same figure, with appropriate title and labellings. fx:(x), fx(x^2), fx(x^3),fx(x^4),fx(x^5),fx(x^6)


```python
import matplotlib.pyplot as plt

def plotfunction(x):
  y1=[i**1 for i in x]
  y2=[i**2 for i in x]
  y3=[i**3 for i in x]
  y4=[i**4 for i in x]
  y5=[i**5 for i in x]
  y6=[i**6 for i in x]
  plt.title('subplots')
  plt.subplot(2,3,1)
  plt.plot(x,y1,'ro--')
  plt.xlabel('X')
  plt.ylabel('Y')
  plt.title('X vs X')
  plt.grid()

  plt.subplot(2,3,2)
  plt.plot(x,y2,'r+--')
  plt.xlabel('X')
  plt.ylabel('X^2')
  plt.title('X vs X^2')
  plt.grid()

  plt.subplot(2,3,3)
  plt.plot(x,y3,'g<--')
  plt.xlabel('X')
  plt.ylabel('X^3')
  plt.title('X vs X^3')
  plt.grid()

  plt.subplot(2,3,4)
  plt.plot(x,y4,'b+--')
  plt.xlabel('X')
  plt.ylabel('X^4')
  plt.title('X vs X^4')
  plt.grid()

  plt.subplot(2,3,5)
  plt.plot(x,y5,'c<--')
  plt.xlabel('X')
  plt.ylabel('X^5')
  plt.title('X vs X^5')
  plt.grid()

  plt.subplot(2,3,6)
  plt.plot(x,y6,'r+--')
  plt.xlabel('X')
  plt.ylabel('X^6')
  plt.title('X vs X^6')
  plt.grid()

  plt.tight_layout()
  plt.show()

def main():
  x=[1,2,3,4]
  plotfunction(x)
main()
```

Given a set of data points 4, 5 , 5 , 7, 6 , 8 , 8 , 9 , 4 , 4 , 4 , 4 , 5 , 5 , 7 , 7 , 8 , 8 , 8. using python plot the frequency of each unique number using histogram. the title of the plot should be plotng histogram. use x lim function to provide the lower limit of x axis as (minimum value - 4) and upper limit of x axis will be (max +4). save the histogram. save the bin range as starting range minimum of data till ending range maximum of data with a width of 5 each

```python
import matplotlib.pyplot as plt
data=[4,5,5,7,6,8,8,9,4,4,4,4,5,5,7,7,8,8,8]
w=5
plt.hist(data,bins=3, edgecolor='black')
plt.xlabel('values')
plt.ylabel('frequency')
plt.title('plotting histogram')
plt.xlim(min(data)-4,max(data)+4)
plt.savefig('histogram.png')
plt.show()
```
